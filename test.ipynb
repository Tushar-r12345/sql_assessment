```python
import pandas as pd
import numpy as np

np.random.seed(42)

n_rows = 100

df = pd.DataFrame({
    'date_timestamp': pd.date_range(start='2022-01-01', end='2023-12-31', periods=n_rows),
    'user_id': np.random.randint(1, 51, size=n_rows), 
    'transaction_id': [f'TXN{i:010d}' for i in range(1, n_rows + 1)],
    'transaction_amount': np.random.uniform(10, 1000, size=n_rows).round(2)
})

df = df.sort_values('date_timestamp').reset_index(drop=True)
```

```python
df.to_excel('df_vals.xlsx')
```

```python
df['date_hour'] = df['date_timestamp'].dt.floor('H')
rouped_df = df.groupby('date_hour').agg(
    total_transactions=('transaction_id', 'count'),
    unique_users=('user_id', 'nunique'),
    total_transaction_amount=('transaction_amount', 'sum')
).reset_index()
```
